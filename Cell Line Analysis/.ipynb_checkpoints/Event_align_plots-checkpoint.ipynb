```python
import os, sys, time
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams['pdf.fonttype'] = 42
```

```python
def ionic_from_eventalign(file_path, ref, begin, end):

    #Define the range of positions we want to establish ionic current measurements for
    position_range = set(range(begin, end+1))
    
    #Read file and eliminate headers produced by eventalign
    ivtFile = open(file_path, 'r')
    ivt_headers = ivtFile.readline().strip().split()
    
    #Dictionary that will contain ivt current level information
    ionic_current_dict = {}
    
    #Read each line of event align and attribute the current at that position to the dictionary list
    for line in ivtFile:
        line = line.strip().split()
        position = int(line[1])
        
        #Only consider positions on the right contig in the bounding range begin-end
        if line[0] != ref:
            continue
        if position not in position_range:
            continue
        
        current = float(line[6])
        if position not in ionic_current_dict:
            ionic_current_dict[position] = {'current':[]}
        ionic_current_dict[position]['current'].append(current)
    
    return ionic_current_dict
```

```python
HepG2_ivtKmerDict = ionic_from_eventalign("HepG2_IVT_rep1.MCM5-35424407.hg38v10.eventalign.txt", 'chr22', 35424403, 35424411)
NTERA_ivtKmerDict = ionic_from_eventalign("NTERA_IVT_rep2.MCM5-35424407.hg38v10.eventalign.txt", 'chr22', 35424403, 35424411)
SHSY5Y_ivtKmerDict = ionic_from_eventalign("SH-SY5Y_undiff_IVT_rep1.MCM5-35424407.hg38v10.eventalign.txt", 'chr22', 35424403, 35424411)
A549_ivtKmerDict = ionic_from_eventalign("A549_IVT_rep1.MCM5-35424407.hg38v10.eventalign.txt", 'chr22', 35424403, 35424411)
HeLa_ivtKmerDict = ionic_from_eventalign("HeLa_IVT_rep3.MCM5-35424407.hg38v10.eventalign.txt", 'chr22', 35424403, 35424411)
```

```python
begin = 35424403
end = 35424411
sns.set(rc={'figure.figsize':(10.5,2)})
for position in HeLa_ivtKmerDict:
    if position >= begin and position <= end:
        plt.clf()
        ivt_ionic_currents = HepG2_ivtKmerDict[position]['current']
        sns.kdeplot(ivt_ionic_currents, shade=False, linewidth=1, alpha=0.75, label = 'HepG2', color = '#4D73B1')

        ivt_ionic_currents = NTERA_ivtKmerDict[position]['current']
        sns.kdeplot(ivt_ionic_currents, shade=False, linewidth=1, alpha=0.75, label = 'NTERA', color = '#DE8452')
        
        ivt_ionic_currents = SHSY5Y_ivtKmerDict[position]['current']
        sns.kdeplot(ivt_ionic_currents, shade=False, linewidth=1, alpha=0.75, label = 'SH-SY5Y', color = '#55A868')
        
        ivt_ionic_currents = A549_ivtKmerDict[position]['current']
        x = sns.kdeplot(ivt_ionic_currents, shade=False, linewidth=1, alpha=0.75, label = 'A549', color = '#C54F53')
        
        ivt_ionic_currents = HeLa_ivtKmerDict[position]['current']
        sns.kdeplot(ivt_ionic_currents, shade=False, linewidth=1, alpha=0.75, label = 'HeLa', color = '#8173B5')

        x.set(xlabel="Event Mean (pA)", ylabel="KDE", title=f"chr22:{position}", xlim=(55, 160))

        plt.plot()
        plt.legend()
        
        #Save figures if executed
#        x.get_figure().savefig(f'{position}.pdf')
```
